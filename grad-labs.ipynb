```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simple function
def f(x):
    return 3 * x ** 2 - 4 * x + 5
x = np.arange(start=-10, stop=10, step=0.1)
y = f(x)
plt.plot(x, y)
plt.show()
```

```python
# Derivitive
# derivitive by formula
def derivitive_by_formula(x):
    return 6 * x - 4

# derivitive by definition (f(x + h) - f(x))/h [https://en.wikipedia.org/wiki/Derivative]
def derivitive_by_definition(x, h):
    return (f(x + h) - f(x)) / h

x_at = -5
h = 0.0001
print(f"Derinitive of y at x = {x_at}")
print(f"derivitive by formula: {derivitive_by_formula(x_at)}")
print(f"derivitive by definition: {derivitive_by_definition(x_at, h)}")
x_at = 5
h = 0.0001
print(f"Derinitive of y at x = {x_at}")
print(f"derivitive by formula: {derivitive_by_formula(x_at)}")
print(f"derivitive by definition: {derivitive_by_definition(x_at, h)}")
x_at = 2/3
h = 0.0001
print(f"Derinitive of y at x = {x_at}")
print(f"derivitive by formula: {derivitive_by_formula(x_at)}")
print(f"derivitive by definition: {derivitive_by_definition(x_at, h)}")
```
